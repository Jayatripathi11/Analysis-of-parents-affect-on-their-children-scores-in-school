GENDER  DISTRIBUTION

```python
plt.figure(figsize=(3,4),dpi=200)
ax=sns.countplot(data=df,x="Gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# from the above chart we have analysed that number of females is more than males
```

IMPACT OF PARENT'S EDUCATION ON STUDENTS MARKS

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
gb
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relationship between parents education and marks of students")
plt.show()
```

```python
#From the above we have analysed that students whose parents have higher education have scored better marks than students of  parents with comparitively lower education
```

IMPACT OF  PARENT'S MARITAL STATUS ON STUDENTS MARKS

```python
gb1=df.groupby("ParentMaritalStatus").agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
```

```python
gb1
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Relationship between parents marital status and marks of students")
plt.show()
```

```python
# from the above chart we have analysed that parent's marital status has negligible impact on their scores
```

DISTRIBUTION OF ETHNIC GROUPS

```python
df['EthnicGroup'].unique()
```

```python
groupA=df.loc[(df['EthnicGroup']=='group A')].count()
groupB=df.loc[(df['EthnicGroup']=='group B')].count()
groupC=df.loc[(df['EthnicGroup']=='group C')].count()
groupD=df.loc[(df['EthnicGroup']=='group D')].count()
groupE=df.loc[(df['EthnicGroup']=='group E')].count()
l=['group A','group B','group C','group D','group E']
mlist=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
print(mlist)
plt.legend(l,title="Ethinicity distribution",loc="best")
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.show();
       
```

```python
#from the above data we have analysed that there are more children of ethinicity C 
```
